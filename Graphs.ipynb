```python
import matplotlib as mt
import pandas as pd
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab as plt
from scipy import signal
import scipy
import time

data = pd.read_csv("ftp_or3.csv")
data = data.convert_objects(convert_numeric=True)

pct = data.C_TRQENG.pct_change()
nlarge = pct.nlargest(5)

ind = nlarge.index[0]
high = ind + 100 
low = ind - 100
window = data.C_TRQENG[low:high]
window = window.convert_objects(convert_numeric=True)
window.plot()
```

```python
ind = nlarge.index[3]
high = ind + 50
low = ind - 50
torq = data.C_TRQENG[low:high]
co2 = data.E_CO2D[low:high]
fig, ax1 = plt.subplots()

lines = ax1.plot(torq)
ax1.set_ylabel('Torq', color='r')

plt.setp(lines, color='r', linewidth=1.0)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
        
ax2 = ax1.twinx()
ax2.plot(co2)

#####

ind = nlarge.index[2]
high = ind + 50
low = ind - 50
torq = data.C_TRQENG[low:high]
co2 = data.E_CO2D[low:high]
fig, ax1 = plt.subplots()

lines = ax1.plot(torq)
ax1.set_ylabel('Torq', color='r')

plt.setp(lines, color='r', linewidth=1.0)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
        
ax2 = ax1.twinx()
ax2.plot(co2)

#####

ind = nlarge.index[1]
high = ind + 50
low = ind - 50
torq = data.C_TRQENG[low:high]
co2 = data.E_CO2D[low:high]
fig, ax1 = plt.subplots()

lines = ax1.plot(torq)
ax1.set_ylabel('Torq', color='r')

plt.setp(lines, color='r', linewidth=1.0)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
        
ax2 = ax1.twinx()
ax2.plot(co2)



#####

ind = nlarge.index[0]
high = ind + 50
low = ind - 50
torq = data.C_TRQENG[low:high]
co2 = data.E_CO2D[low:high]
fig, ax1 = plt.subplots()

lines = ax1.plot(torq)
ax1.set_ylabel('Torq', color='r')

plt.setp(lines, color='r', linewidth=1.0)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
        
ax2 = ax1.twinx()
ax2.plot(co2)



#####

ind = nlarge.index[2]
high = ind + 50
low = ind - 50
torq = data.C_TRQENG[low:high]
co2 = data.E_CO2D[low:high]
fig, ax1 = plt.subplots()

torq = torq/torq.loc[torq.abs().idxmax()].astype(np.float64)
co2 = co2/co2.loc[co2.abs().idxmax()].astype(np.float64)

shift =  numpy.correlate(co2, torq)
shift = shift[0] - shift[0] * 2 
lines = ax1.plot(torq)
ax1.set_ylabel('Torq', color='r')

plt.setp(lines, color='r', linewidth=1.0)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
        
ax2 = ax1.twinx()
ax2.plot(co2.shift(shift))
dfco2 = co2.diff()
dftorq = torq.diff()
dftorq.to_csv("torq.csv")

dfco2.to_csv("co2.csv")
```
