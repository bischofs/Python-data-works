```python
import matplotlib as mt
import pandas as pd
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab as plt
from scipy import signal
import scipy
import time
%matplotlib inline

data = pd.read_csv("ftp_or3.csv")
data = data.convert_objects(convert_numeric=True)
data = data.dropna()
pct = data.C_TRQENG.pct_change()
nlarge = pct.nlargest(5)
products = []
sums = []


ind = nlarge.index[4]
high = ind + 100
low = ind - 100


for i in range(-1, -20, -1):
    shifted_species = data.E_NOXD.shift(i)
    shifted_species = shifted_species.dropna()

    for spec, torq in zip(shifted_species[low:high], data.C_TRQENG[low:high]):
        products.append(spec * torq)
    
    
    sum_result = (sum(products))
   
    sums.insert(i,sum_result)
    

    
data.E_NOXD * data.C_TRQENG
#plt.plot(sums)
data.C_TRQENG
```

```python
torq = data.C_TRQENG[low:high]
co2 = data.E_NOXD[low:high]
fig, ax1 = plt.subplots()

lines = ax1.plot(torq)
ax1.set_ylabel('Torq', color='r')

plt.setp(lines, color='r', linewidth=1.0)
for tl in ax1.get_yticklabels():
    tl.set_color('r')
        
ax2 = ax1.twinx()
ax2.plot(co2)
```
